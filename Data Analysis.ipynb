```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data= pd.read_csv('Laptops under 40000.csv')
```

```python
data.head
```

Find the product having the highest rating

```python
data[data['Ratings']== max(data['Ratings'])]
```

```python
data.sort_values(by='Ratings', ascending=False).head(5)
```

```python
plt.figure(figsize=(15,7))
plt.bar(x=data['product_name'], height=data['Ratings'])
plt.title('Ratings of Laptop', fontsize=30)
plt.xlabel('Laptops name', fontsize=30)
plt.ylabel('Ratings', fontsize=30)
plt.xticks(rotation=90)
plt.show()
```

Affordable laptops

```python
data.sort_values(by='current_price', ascending=True).head(5)
```

```python
plt.figure(figsize=(30,7))
plt.bar(x=data['product_name'], height=data['current_price'])
plt.title('Ratings of Laptop', fontsize=30)
plt.xlabel('Laptops name', fontsize=30)
plt.ylabel('Current price', fontsize=30)
plt.xticks(rotation=270)
plt.show()
```
